```python
import pandas as pd
import numpy as np
```

```python
inspections = pd.read_csv("../Data/chapter_06_working_with_text_data/chicago_food_inspections.csv")
inspections
```

```python
#remove left whitespaces and right whitespaces
inspections["Name"] = inspections["Name"].str.strip()
```

```python
inspections
```

```python
#removing white spaces from all the columns
for column in inspections.columns:
    inspections[column] = inspections[column].str.strip()
inspections
```

```python
#using title to capitalise only the first alphabet
for column in inspections.columns:
    inspections[column] = inspections[column].str.title()
inspections
```

```python
#removing na rows in risk column
inspections = inspections.dropna(subset=['Risk'])
```

```python
#Unique values in Risk Column
inspections['Risk'].unique()
```

```python
#Replace All with Risk 4 (Extreme)
inspections = inspections.replace(to_replace='All',value='Risk 4 (Extreme)')
inspections['Risk'].unique()
```

```python
#Slicing
val = inspections['Risk'].iloc[0]
idx = 0
for a in val:
    if a == "(":
        idx+=1
        break
    idx+=1
inspections['Risk'].str[8:].str.replace(")","")
```

```python
has_pizza = inspections['Name'].str.lower().str.contains('pizza')
inspections[has_pizza]
```
