```python
import numpy as np
import pandas as pd
```

```python
emp = pd.read_csv('../Data/chapter_05_filtering_a_dataframe/employees.csv',parse_dates=["Start Date"])
```

```python
emp.info()
```

```python
#converting datatype with astype
emp['Mgmt']=emp['Mgmt'].astype(bool)
```

```python
emp.info()
```

```python
#fill na to be zero and convert to int
emp['Salary'] = emp['Salary'].fillna(0).astype(int)
```

```python
emp.info()
```

```python
#setting value of gender column to be category as it can only be Male or Female
emp['Gender'] = emp['Gender'].astype('category')
```

```python
emp.info()
```

```python
emp['Team'] = emp['Team'].astype('category')
```

```python
emp.info()
```

```python
#comparing a series with a constant value
emp['First Name'] == 'Maria'
```

```python
#getting the dataframe rows with Maria as first name using Boolean series filter
emp[emp['First Name'] == 'Maria']
```

```python
#storing boolean series in a variable
first_name_maria = (emp['First Name'] == 'Maria')
first_name_maria
```

```python
emp[first_name_maria]
```

```python
#non finance teams filter
non_finance = (emp['Team'] != 'Finance')
emp[non_finance]
```

```python
mgmt_team = (emp['Mgmt'] == True)
emp[mgmt_team]
```

```python
#highest earner filter > 100000
high_earner = (emp['Salary'] > 100000)
emp[high_earner]
```

```python
#filtering  by multiple condition
#and condition
is_female = (emp['Gender'] == 'Female')
is_business_dev = (emp['Team'] == "Business Dev")
```

```python
emp[is_female & is_business_dev]
```

```python
is_manager = (emp['Mgmt'] == True)
emp[is_female & is_business_dev & is_manager]
```

```python
#or operator
is_earning_below_40k = (emp['Salary'] < 40000)
after_2015 = (emp['Start Date'] >= '2015-01-01')
emp[is_earning_below_40k | after_2015]
```

```python
salary_not_zero = (emp['Salary'] != 0)
emp[(is_earning_below_40k | after_2015) & salary_not_zero ]
```

```python
#not operator
emp[~is_manager]
```

```python
#using isin method 
sales = emp['Team'] == 'Sales'
finance = emp['Team'] == 'Finance'
legal = emp['Team'] == 'Legal'
```

```python
all_star_teams = ['Sales','Finance','Legal']
on_all_star_teams = emp['Team'].isin(all_star_teams)
```

```python
emp[sales | finance | legal]
```

```python
emp[on_all_star_teams]
```

```python
#between operator lowerbound is inclusive and outer bound is exclusive
between_80k_and_90k = emp['Salary'].between(left=80000,right=90000)
emp[between_80k_and_90k]
```

```python
eighties_folk = emp['Start Date'].between(left='1980-01-01',right='1990-01-01')
emp[eighties_folk]
```

```python
#isnull isnot null
team_null = emp['Team'].isnull()
team_null
```

```python
emp[team_null]
```

```python
start_date_null = emp['Start Date'].isnull()
emp[start_date_null]
```

#### Dealing with Null Values

#### Dealing with duplicates

```python
#duplicated method
emp['Team'].duplicated()
```

```python
emp['Team'].duplicated(keep='first') #by default value is first
```

```python
emp['Team'].duplicated(keep='last')
```

```python
#drop duplicates method
emp.drop_duplicates()
```

```python
emp.drop_duplicates(subset=['Team']).dropna(subset='Team')
```

```python
emp.drop_duplicates(subset=['Gender','Team']).dropna().sort_values(by=['Gender','Team'])
```
